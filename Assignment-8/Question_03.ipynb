```python
import pandas as pd 


# Create first DataFrame
df1 = pd.DataFrame({

    # Common column used for merging
    "id":[1,2,3,4],

    # Names of students/users
    "Name":["Raj","Suresh","Aman","Ravi"],

    # Skills column
    "skills":["C++","Python","Java","ML"]
})


# Create second DataFrame
df2 = pd.DataFrame({

    # Common column used for merging
    "id":[1,2,3,4],

    # Another set of names
    "Name":["Neha","Kanak","Priya","Nidhi"],

    # Skills column
    "skills":["C","JavaScript","ML","AI"]
})

df3=pd.DataFrame({
     
    # Common column used for merging
    "id":[1,2,3,4],

    # Names of students/users
    "Name":["Gunjan","Muskan","Vivek","Raina"],

    # Skills column
    "skills":["Database","Computer Network","Cotlin","HTML"]

})

# Print first DataFrame
print(df1)

# Print second DataFrame
print(df2)

# Print third DataFrame
print(df3)
```

```python
# Vertically concatenate two DataFrames

df4 = pd.concat([df1, df2], axis=0)

print(df4)


# Merge the resulting DataFrame with df1

merge_data = df4.merge(df1, on="id", how="left")

print(merge_data)
```

```python
# Explain the primary differences between df.join() and pd.merge()

# 1) pd.merge()

# merge() is used to combine DataFrames using one or more common columns. It works similar to SQL joins in databases.

# Features of merge():--

# Combines DataFrames based on common column values.
# Supports different types of joins:
# Inner Join
# Left Join
# Right Join
# Outer Join
# It Can merge using single or multiple keys (columns).


# 2) df.join()

# join() is mainly used to combine DataFrames based on their indexes.

# Features of join():--

# Joins DataFrames using index values by default.
# Simpler syntax compared to merge().
# Mostly used when indexes are meaningful.

```
