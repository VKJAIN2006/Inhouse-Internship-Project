```python
import pandas as pd 


# Create first DataFrame
df1 = pd.DataFrame({

    # Common column used for merging
    "id":[1,2,3,4],

    # Names of students/users
    "Name":["Raj","Suresh","Aman","Ravi"],

    # Skills column
    "skills":["C++","Python","Java","ML"]
})


# Create second DataFrame
df2 = pd.DataFrame({

    # Common column used for merging
    "id":[1,2,3,4],

    # Another set of names
    "Name":["Neha","Kanak","Priya","Nidhi"],

    # Skills column
    "skills":["C","JavaScript","ML","AI"]
})


# Print first DataFrame
print(df1)

# Print second DataFrame
print(df2)

```

```python
# Perform inner merge on common column 'id'

# Only matching ids from both DataFrames will be merged
res1 = df1.merge(df2, on="id", how="inner")


# Print merged DataFrame
print(res1)
```

```python
# >Perform a left join of df1 and df2 on the 'ID' column. Explain how missing values are handled in the resulting DataFrame. 
# Right Join and Index-Based Join.

# Perform left merge on common column 'id'
res1 = df1.merge(df2, on="id", how="left")

# Perform right merge on common column 'id'
res2 = df1.merge(df2, on="id", how="right")

print(res1)
print(res2)


# How Missing Values Are Handled?

# In Pandas:
# When a matching value is not found during merge,
# Pandas fills missing data with: NaN (Not a Number)
```

```python
# >Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. 
# Compare the results. Merging with Multiple Keys.


# Right Join
right_join = df1.merge(df2, on="id", how="right")

print("Right Join\n")

print(right_join)


# Index-Based Join
join_data = df1.join(df2, lsuffix="_df1", rsuffix="_df2")

print("\nIndex Based Join\n")

print(join_data)

#Comparing the results

# 1. merge() with Multiple Keys:-- merging with multiple keys means,both columns must match together for rows to merge.
# 2. join() with multiple keys:--- join() uses row indexes not column values
```
