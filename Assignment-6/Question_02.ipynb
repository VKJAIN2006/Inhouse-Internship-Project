```python
## Rock Paper Scissor Game ##
import random
I=["rock","paper","scissor"]
uscore=0
cscore=0

while True:
        print("""
    1. Play Game
    2. Exit
    """)
        c=int(input("User waants to play or not"))
               

        if(c==1):
          

          for i in range(1,6):
                
                uch=input("Enter user choice  :")
                if uch not in I:
                    print("Invalid Choice!")
                    continue
                cch=random.choice(I)
                print("User choice is : ",uch)
                print("Computer choice is : ",cch)
                if(uch==cch):
                 print("Match Draw")
                 uscore+=1
                 cscore+=1
                elif((uch=="rock" and cch=="scissor")or (uch=="paper" and cch=="rock") or (uch=="scissor" and cch=="paper")):
                   uscore+=2
                   print("User Wins")
                else:
                   cscore+=2
                   print("Computer wins")

          print("\n===== Final Scores =====")

          print("User Score :", uscore)
          print("Computer Score :", cscore)

        
        
        elif(c==2):
            break
     # Final Winner
        if uscore > cscore:
            print("User Wins The Game")

        elif cscore > uscore:
            print("Computer Wins The Game")

        else:
            print("Game Draw")

     


```
