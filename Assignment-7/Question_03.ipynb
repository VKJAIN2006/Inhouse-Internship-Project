```python
# 3)Data iteration
# 	Different ways to iterate over rows in Pandas Dataframe 
# 	Selecting rows in pandas DataFrame based on conditions 
# 	Select any row from a Dataframe using iloc[]
# 	Limited rows selection with given column
# 	Drop rows from the dataframe based on certain condition applied on a column 
# 	Insert row at given position in Pandas Dataframe 
# 	Create a list from rows in Pandas dataframe 
```

```python
import pandas as pd

data = {

    "ID": [101, 102, 103, 104, 105, 106, 107],

    "Name": ["Vivek", "Keshav", "Nidhi", "Priya", "Neha", "Rahul", "Aman"],

    "Age": [20, 21, 22, 20, 23, 24, 21],

    "Branch": ["CSE", "IT", "EE", "AI", "CSE", "ME", "IT"],

    "City": ["Jaipur", "Delhi", "Kota", "Alwar", "Jaipur", "Mumbai", "Delhi"],

    "Marks": [85, 78, 91, 88, 76, 67, 95]

}

df = pd.DataFrame(data)

print(df)
```

```python
# Different ways to iterate over rows in Pandas Dataframe 
for row in df.itertuples():

    print(row)
```

```python
# Selecting rows in pandas DataFrame based on conditions 
row= df[df['Marks']>80]
print(row)
```

```python
# Select any row from a Dataframe using iloc[]
print(df.iloc[0:3])
```

```python
# Limited rows selection with given column
newdata = df.loc[0:2, ['Name', 'Marks']]

print(newdata)
```

```python
# Drop rows from the dataframe based on certain condition applied on a column 

newdata = df.drop(df[df['Marks'] < 80].index)
print(newdata)
```

```python
# Insert row at given position in Pandas Dataframe 
# New row data
new_row = {

    "ID": 108,

    "Name": "Priya",

    "Age":20,

    "Branch":"CSE",

    "City":"Ajmer",

    "Marks": 88

}

# Insert row at position 1

upper = df.iloc[:1]

lower = df.iloc[1:]

newdata = pd.concat([upper, pd.DataFrame([new_row]), lower], ignore_index=True)

print("\nDataFrame After Inserting Row\n")

print(newdata)
```

```python
# Create a list from rows in Pandas dataframe 
lst = df.values.tolist()

print(lst)
```
