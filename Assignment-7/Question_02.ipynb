```python
import pandas as pd

# 1. Make a Pandas DataFrame with a Two-Dimensional Python List

data1 = [
    [101, "Vivek", "CSE"],
    [102, "Keshav", "IT"],
    [103, "Nidhi", "EE"]
]

df1 = pd.DataFrame(data1)

print("DataFrame from Two-Dimensional List\n")

print(df1)
```

```python
# 	Create DataFrame from Python dict 
data_set={
    "Languages":["Python","C++","Java","JavaScript","C#"],
    "Scores":[90,90,80,100,89],
    "Domains":["Data Scientist","Gamer","Developer","Web Development","Game Development"]
   }
df=pd.DataFrame(data_set)
print("\nDataFrame from Dictionary\n")
print(df)
```

```python
# 	Create Pandas dataframe using list of lists

lst=[
    ["Hawa Mahal","Gaurav Tower","Hight Street"],
    ["Dal Bati","Ghevar","Dosa"],
    ["Govind Dev ji","Galta ji","Moti Dungri"],
    ["Amer Fort","Nahargarh Fort","Jaigarh Fort"]
] 
df1=pd.DataFrame(lst)
print(df1)
```

```python
# 4. Create Pandas DataFrame using List of Tuples

data4 = [
    ("Vivek", 85),
    ("Keshav", 78),
    ("Nidhi", 91)
]

df4 = pd.DataFrame(data4, columns=["Name", "Marks"])

print("\nDataFrame from List of Tuples\n")

print(df4)
```

```python
# 5. Create Pandas DataFrame from List of Dictionaries

data5 = [
    {"Name": "Vivek", "Age": 20, "City": "Jaipur"},
    {"Name": "Keshav", "Age": 21, "City": "Delhi"},
    {"Name": "Nidhi", "Age": 22, "City": "Kota"}
]

df5 = pd.DataFrame(data5)

print("\nDataFrame from List of Dictionaries\n")

print(df5)
```
