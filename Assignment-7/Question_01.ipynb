```python
# 1) Practise Pandas Series
# 	Create a Pandas Series from Dictionary 
# 	Create a Pandas Series from Lists 
# 	Access the elements of a Series in Pandas  
```

```python
import pandas as pd

# Pandas Series from Dictionary

student_data = {
    "Name": "Vivek Jain",
    "Section": "F",
    "Branch": "CSE",
    "Roll_no": "123456789",
    "Clg_ID": "B24XXXX"
}

series1 = pd.Series(student_data)

print("Series from Dictionary\n")

print(series1)

print("\nAccessing Elements")

print(series1["Name"])

print(series1["Clg_ID"])
```

```python
# Pandas Series from List

lst = ["Jaipur", "Ajmer", "HawaMahal", 25, 10, 1998, 2007]

series2 = pd.Series(lst)

print("\nSeries from List\n")

print(series2)

print("\nAccessing Elements")

print(series2[0])

print(series2[5])
```
