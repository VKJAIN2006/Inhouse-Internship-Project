```python
import numpy as np
```

```python


# Create array
arr = np.array([10, 20, 30, 40, 50])

# Print original array
print("Original Array:")
print(arr)

# Reverse the array
reverse_arr = arr[::-1]

# Print reversed array
print("Reversed Array:")
print(reverse_arr)
```
