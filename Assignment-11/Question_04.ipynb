```python
import numpy as np
```

```python
# Create a 2D NumPy array
arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
# Print original array
print("Original Array:")
print(arr)
```

```python
# Get maximum value from array
print("\nMaximum Value:")
print(arr.max())

# Get minimum value from array
print("\nMinimum Value:")
print(arr.min())
```

```python
# Find number of rows and columns
rows = arr.shape[0]
cols = arr.shape[1]

print("\nNumber of Rows:", rows)
print("Number of Columns:", cols)
```

```python
# Select all elements one by one
print("\nAll Elements:")

for i in range(rows):
    for j in range(cols):
        print(arr[i][j])
```

```python
# Select specific element
print("\nSpecific Element at row 1 column 2:")
print(arr[1][2])
```

```python
# Find sum of all elements using for loop
total = 0

for i in range(rows):
    for j in range(cols):
        total += arr[i][j]

print("\nSum of All Elements:")
print(total)
```

```python
# Create another array for arithmetic operations
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("\nSecond Array:")
print(arr2)
```

```python
# Addition of arrays
print("\nAddition:")
print(arr + arr2)
```

```python
# Subtraction of arrays
print("\nSubtraction:")
print(arr - arr2)
```

```python
# Multiplication of arrays
print("\nMultiplication:")
print(arr * arr2)
```

```python
# Division of arrays
print("\nDivision:")
print(arr / arr2)
```
