```python
# Create first array
arr1 = np.array([3, 4])

# Create second array
arr2 = np.array([1, 0])

# Print first array
print(arr1)

# Print second array
print(arr2)

# Calculate average of corresponding elements
avg = (arr1 + arr2) / 2

# Print average array
print("Average of NumPy arrays:\n", avg)
```

```python
import numpy as np
from scipy import stats   # Import stats module for mode

# Create first 2D array
arr1 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# Create second 2D array
arr2 = np.array([
    [15, 25, 35],
    [45, 55, 65]
])

# Print arrays
print("First Array:")
print(arr1)

print("\nSecond Array:")
print(arr2)
```

```python
# Calculate Mean (Average)
print("\nMean of First Array:")
print(np.mean(arr1))

print("\nMean of Second Array:")
print(np.mean(arr2))
```

```python
# Calculate Median
print("\nMedian of First Array:")
print(np.median(arr1))

print("\nMedian of Second Array:")
print(np.median(arr2))
```

```python
# Calculate Mode
print("\nMode of First Array:")
print(stats.mode(arr1, axis=None))

print("\nMode of Second Array:")
print(stats.mode(arr2, axis=None))
```
