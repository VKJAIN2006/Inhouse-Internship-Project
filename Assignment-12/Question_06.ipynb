```python
import numpy as np

# Create a 4×4 array
arr = np.arange(1, 17).reshape(4, 4)

print("Original Array:")
print(arr)

# Extract odd rows and even columns
result = arr[::2, 1::2]

print("\nOdd Rows and Even Columns:")
print(result)
```

```python
# > Slice the first two rows and first two columns of second set from a 4×4×3 array

# Create a 4×4 array
arr = np.arange(1, 49).reshape(4, 4,3)

print("Original Array:")
print(arr)

result = arr[1, 0:2, 0:2]

print("\nRequired Slice:")
print(result)
```
