```python
import numpy as np

arr = np.array([1, 2, 3, 45, 60, 20])

print("Array:", arr)
print("Shape:", arr.shape)      # Number of rows and columns
print("Dimension:", arr.ndim)   # Number of dimensions
print("Data Type:", arr.dtype)  # Type of elements
print("Item Size:", arr.itemsize, "bytes")  # Size of one element
print("Size:", arr.size)        # Total number of elements
```
