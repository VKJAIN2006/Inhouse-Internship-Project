```python
# > Create a 4×4×3 array and extract value 
# at its second set, first row and last column 
```

```python

# Create a 4×4×3 array
arr = np.arange(48).reshape(4, 4, 3)

print(arr)

# Extract value at:
# second set -> index 1
# first row  -> index 0
# last column -> index 2
value = arr[1, 0, 2]

print("Required value:", value)
```
