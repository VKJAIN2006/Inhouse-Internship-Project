```python
# Import pandas library
import pandas as pd


# Convert string dates into datetime format
dates = pd.to_datetime([

    "2026-05-21",
    "2026-06-15",
    "2026-07-10"
])


# Print complete datetime series
print(dates)


# Extract and print year from dates
print(dates.year)


# Extract and print complete date only
print(dates.date)


# Extract and print month number
print(dates.month)


# Extract and print day number
print(dates.day)


# Extract and print day names
print(dates.day_name())


# Extract and print month names
print(dates.month_name())
```
