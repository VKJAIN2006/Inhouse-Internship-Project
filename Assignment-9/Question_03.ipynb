```python
# Import pandas library for data analysis
import pandas as pd

# Read CSV file and store it in dataframe
df = pd.read_csv("practise.csv")

# Display first 5 rows of dataset
print(df.head())
```

```python
# Display all column names present in dataset
print(df.columns)
```

```python
# Check total number of null (missing) values in each column
print(df.isnull().sum())
```

```python
# Display dataset information such as:
# total rows, column names, data types, and non-null values
print(df.info())
```

```python
# Generate statistical summary of numerical columns
# It shows count, mean, standard deviation, minimum,
# maximum, and quartile values
print(df.describe())
```

```python
# Fill missing values in Date column using forward fill method
df["Date"] = df["Date"].fillna(method='ffill')

# Fill missing values in Calories column using forward fill method
df["Calories"] = df["Calories"].fillna(method='ffill')

# Check whether null values are removed or not
print(df.isnull().sum())
```

```python
# Convert Date column into datetime format
# Invalid date values will be converted into NaT (missing datetime)
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')

# Convert Date column using specific date format: month-day-year
df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%Y")
```

```python
# Display maximum value from Maxpulse column
print(df['Maxpulse'].max())

# Display minimum value from Maxpulse column
print(df['Maxpulse'].min())

# Display maximum value from Calories column
print(df['Calories'].max())

# Display minimum value from Calories column
print(df['Calories'].min())
```
