```python
# 1) Replace Nan with 0 and 
# Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]]
```

```python
import pandas as pd
import numpy as np
```

```python
import numpy as np

# Create 2D array
arr = np.array([

    [6, -8, 73, -110],

    [np.nan, -8, 0, 94]

])

print("Original Array:\n")

print(arr)


# Replace NaN with 0
new_arr = np.nan_to_num(arr, nan=0)

print("\nAfter Replacing NaN with 0:\n")

print(new_arr)


# Interchange rows and columns (Transpose)
transpose_arr = new_arr.T

print("\nAfter Interchanging Rows and Columns:\n")

print(transpose_arr)
```
